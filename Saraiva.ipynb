```python
import pandas as pd
```

```python
import requests as r
```

```python
from pandas.io.json import json_normalize
```

```python
df = r.get('https://www.googleapis.com/books/v1/volumes?q=inpublisher:Saraiva+Educação&orderBy=newest')
```

```python
lista = df.json()
```

```python
lista['items'][1]['accessInfo']['pdf']['isAvailable']
```

```python
idd = []
title = []
isbn = []
authors = []
dataPublicacao = []
anoPublicacao = []
page = []
salepermition = []
listprice = []
categories = []
publisher = []
canonicalvolume = []
epubavailable = []
pdfbavailable = []
```

```python
for item in range(10): ##Para encontrar e listar os id's
    x = lista['items'][item]['id'] 
    idd = idd+[x]

for item in range(10): ##Para encontrar e listar os titulos
    x = lista['items'][item]['volumeInfo']['title']
    title = title + [x]
    
for item in range(10): ##Para encontrar e listar os ISBn's
    x = lista['items'][item]['volumeInfo']['industryIdentifiers'][0]['identifier']
    isbn = isbn+[x]

for item in range(10): ##Para encontrar e listar os autores
    x = lista['items'][item]['volumeInfo']['authors']
    authors = authors + [x]
    
for item in range(10): ##Para encontrar a data de publicação
    x = lista['items'][1]['volumeInfo']['publishedDate']
    dataPublicacao = dataPublicacao+[x]
    
for item in range(10):##Para separar apenas o ano
    x = dataPublicacao[item][:4]
    anoPublicacao = anoPublicacao+[x]

for item in range(10): ##Para econtrar e listar disponibilidade para vendas
    x = lista['items'][item]['saleInfo']['saleability']
    salepermition = salepermition + [x]
    
for item in range(10): ##Para encontrar e listar o preço
    x = lista['items'][item]['saleInfo']['listPrice']['amount']
    listprice = listprice + [x]
    
for item in range(10): ##Para encontrar e listar as categorias
    x = lista['items'][item]['volumeInfo']['categories']
    categories = categories + [x]
    
for item in range(10): ##Para listar a editora
    x = lista['items'][item]['volumeInfo']['publisher']
    publisher = publisher + [x]
    
for item in range(10): ##Para encontrar o link para o google books
    x = lista['items'][item]['volumeInfo'][ 'canonicalVolumeLink']
    canonicalvolume = canonicalvolume + [x]
    
for item in range(10): ##Para listar a disponibilidade do epub
    x = lista['items'][item]['accessInfo']['epub']['isAvailable']
    epubavailable = epubavailable + [x]
    
for item in range(10): ##Disponibilidade de PDF
    x = lista['items'][item]['accessInfo']['pdf']['isAvailable']
    pdfbavailable = pdfbavailable + [x]
```

```python
booksInfo = {'Idd':idd,'Title':title, 'Isbn':isbn, 'Authors':authors, 'Data Publicação':dataPublicacao, 'Ano Publicação':anoPublicacao,
            'Permissão de venda':salepermition, 'Lista de Preço':listprice,'Categorias':categories, 'Editora':publisher, 'Link google books':canonicalvolume,
            'Epub':epubavailable,'Pdf':pdfbavailable}##criando um dicionario com as listas geradas acima
```

```python
data =pd.DataFrame(booksInfo,index = range(0,10))
```
